```python
import pandas as pd
import os
```

```python
kpis = pd.read_csv("../output/dashboard_kpis.csv")
companies = pd.read_csv("../output/company_insights.csv")
categories = pd.read_csv("../output/category_insights.csv")

kpis.head()
```

```python
dashboard_category = categories.copy()

dashboard_category["Dashboard_Type"] = "Category"

dashboard_category.head()
```

```python
dashboard_company = companies.copy()

dashboard_company["Dashboard_Type"] = "Company"

dashboard_company.head()
```

```python
dashboard_category = categories.copy()

dashboard_category["Dashboard_Type"] = "Category"

dashboard_category.head()
```

```python
import os

os.makedirs("../output", exist_ok=True)

dashboard_company.to_csv(
    "../output/dashboard_company_dataset.csv",
    index=False
)

dashboard_category.to_csv(
    "../output/dashboard_category_dataset.csv",
    index=False
)

kpis.to_csv(
    "../output/dashboard_summary.csv",
    index=False
)

print("Dashboard datasets generated successfully!")
```
