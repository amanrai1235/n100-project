```python
import pandas as pd
import os
```

```python
df = pd.read_csv("../output/financial_metrics_standardized.csv")

df.head()
```

```python
print(df.info())
print(df.describe(include="all"))
```

```python
category_summary = (
    df.groupby("metric_category")
      .agg(
          total_metrics=("metric_type", "count"),
          avg_value=("value_pct", "mean"),
          max_value=("value_pct", "max"),
          min_value=("value_pct", "min")
      )
      .reset_index()
)

category_summary
```

```python
company_summary = (
    df.groupby("company_id")
      .agg(
          metrics_available=("metric_type", "count"),
          average_score=("value_pct", "mean")
      )
      .reset_index()
)

company_summary.head()
```

```python
os.makedirs("../output", exist_ok=True)

category_summary.to_csv(
    "../output/category_summary.csv",
    index=False
)

company_summary.to_csv(
    "../output/company_summary.csv",
    index=False
)

print("Summary reports generated successfully!")
```
