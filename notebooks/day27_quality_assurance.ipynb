```python
import os
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("../database/nifty100.db")
```

```python
tables = pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table';
""", conn)

tables
```

```python
for table in tables["name"]:
    count = pd.read_sql(
        f"SELECT COUNT(*) AS total FROM {table}",
        conn
    )

    print(table)
    print(count)
    print("-"*40)
```

```python
reports = [
    "../reports/valuation_flags.csv",
    "../reports/valuation_summary.xlsx"
]

for file in reports:

    if os.path.exists(file):
        print(file, "Found")
    else:
        print(file, "Missing")
```

```python
valuation = pd.read_csv("../reports/valuation_flags.csv")

valuation.isnull().sum()
```
