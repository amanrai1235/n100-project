```python
candidate_path = "../reports/day21_top_investment_candidates.csv"

df = pd.read_csv(candidate_path)

print(df.shape)
print(df.columns.tolist())

df.head()
```

```python
portfolio = df[
    [
        "company_id",
        "composite_score",
        "quality_score_100",
        "momentum_score_100",
        "normalized_risk_score",
        "risk_category",
        "trend_classification"
    ]
].copy()

portfolio
```

```python
portfolio = portfolio.sort_values(
    by="composite_score",
    ascending=False
).reset_index(drop=True)

portfolio["portfolio_rank"] = portfolio.index + 1

portfolio
```

```python
portfolio["allocation_percent"] = (
    portfolio["composite_score"]
    / portfolio["composite_score"].sum()
) * 100

portfolio["allocation_percent"] = portfolio[
    "allocation_percent"
].round(2)

portfolio
```

```python
def recommendation(score):
    if score >= 70:
        return "Strong Buy"
    elif score >= 50:
        return "Buy"
    elif score >= 30:
        return "Hold"
    else:
        return "Watch"

portfolio["recommendation"] = portfolio[
    "composite_score"
].apply(recommendation)

portfolio
```

```python
portfolio = portfolio[
    [
        "portfolio_rank",
        "company_id",
        "recommendation",
        "allocation_percent",
        "composite_score",
        "quality_score_100",
        "momentum_score_100",
        "normalized_risk_score",
        "risk_category",
        "trend_classification"
    ]
]

portfolio
```

```python
portfolio.to_csv(
    "../reports/day22_portfolio.csv",
    index=False
)

print("Saved Successfully")
```
