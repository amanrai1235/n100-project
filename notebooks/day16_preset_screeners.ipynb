# Day 16 - Preset Financial Screeners

## Objectives

- Load validated financial ratio data
- Build predefined financial screening strategies
- Implement Quality Screener
- Implement Low Debt Screener
- Implement High ROE Screener
- Implement Income Screener
- Implement Balanced Screener
- Generate preset screener reports

```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv(
    "../reports/financial_ratios_day12.csv"
)

print("Financial Ratio Data Loaded Successfully")
print("Shape:", df.shape)

df.head()
```

```python
numeric_cols = [
    "return_on_equity_pct",
    "debt_to_equity",
    "interest_coverage",
    "asset_turnover",
    "dividend_payout_ratio"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(
        df[col],
        errors="coerce"
    )

print(df[numeric_cols].dtypes)
```

```python
latest_df = (
    df.sort_values("year")
      .drop_duplicates(
          subset=["company_id"],
          keep="last"
      )
      .reset_index(drop=True)
)

print("Original Records:", len(df))
print("Latest Company Records:", len(latest_df))

latest_df.head()
```

```python
latest_df = df.copy()

latest_df["year_number"] = pd.to_numeric(
    latest_df["year"].astype(str).str.extract(r"(\d{4})")[0],
    errors="coerce"
)

latest_df = (
    latest_df.sort_values("year_number")
    .drop_duplicates(
        subset=["company_id"],
        keep="last"
    )
    .reset_index(drop=True)
)

print("Original Records:", len(df))
print("Latest Company Records:", len(latest_df))

latest_df[[
    "company_id",
    "year",
    "year_number"
]].head()
```

```python
quality_screener = latest_df[
    (latest_df["return_on_equity_pct"] >= 15) &
    (latest_df["debt_to_equity"] <= 1) &
    (latest_df["interest_coverage"] >= 3)
].copy()

print(
    "Quality Screener Results:",
    len(quality_screener)
)

quality_screener.head(10)
```

```python
low_debt_screener = latest_df[
    latest_df["debt_to_equity"] <= 0.5
].copy()

print(
    "Low Debt Screener Results:",
    len(low_debt_screener)
)

low_debt_screener.head(10)
```

```python
high_roe_screener = latest_df[
    latest_df["return_on_equity_pct"] >= 20
].copy()

print(
    "High ROE Screener Results:",
    len(high_roe_screener)
)

high_roe_screener.head(10)
```

```python
income_screener = latest_df[
    (latest_df["dividend_payout_ratio"] >= 20) &
    (latest_df["return_on_equity_pct"] >= 10)
].copy()

print(
    "Income Screener Results:",
    len(income_screener)
)

income_screener.head(10)
```

```python
balanced_screener = latest_df[
    (latest_df["return_on_equity_pct"] >= 12) &
    (latest_df["debt_to_equity"] <= 1.5) &
    (latest_df["asset_turnover"] >= 0.5)
].copy()

print(
    "Balanced Screener Results:",
    len(balanced_screener)
)

balanced_screener.head(10)
```

```python
preset_summary = pd.DataFrame({
    "preset_name": [
        "Quality Screener",
        "Low Debt Screener",
        "High ROE Screener",
        "Income Screener",
        "Balanced Screener"
    ],
    "records_found": [
        len(quality_screener),
        len(low_debt_screener),
        len(high_roe_screener),
        len(income_screener),
        len(balanced_screener)
    ]
})

preset_summary
```

```python
quality_screener["preset"] = "Quality"
low_debt_screener["preset"] = "Low Debt"
high_roe_screener["preset"] = "High ROE"
income_screener["preset"] = "Income"
balanced_screener["preset"] = "Balanced"
```

```python
all_preset_results = pd.concat(
    [
        quality_screener,
        low_debt_screener,
        high_roe_screener,
        income_screener,
        balanced_screener
    ],
    ignore_index=True
)

print(
    "Total Preset Result Records:",
    len(all_preset_results)
)

all_preset_results.head()
```

```python
quality_screener.to_csv(
    "../reports/day16_quality_screener.csv",
    index=False
)

low_debt_screener.to_csv(
    "../reports/day16_low_debt_screener.csv",
    index=False
)

high_roe_screener.to_csv(
    "../reports/day16_high_roe_screener.csv",
    index=False
)

income_screener.to_csv(
    "../reports/day16_income_screener.csv",
    index=False
)

balanced_screener.to_csv(
    "../reports/day16_balanced_screener.csv",
    index=False
)

preset_summary.to_csv(
    "../reports/day16_preset_summary.csv",
    index=False
)

all_preset_results.to_csv(
    "../reports/day16_all_preset_results.csv",
    index=False
)

print("Day 16 Preset Screener Reports Saved Successfully!")
```
