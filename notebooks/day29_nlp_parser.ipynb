```python
df = pd.read_excel(analysis_path, header=None)

# Second row ko header banao
df.columns = df.iloc[1]

# Pehli 2 rows hata do
df = df.iloc[2:].reset_index(drop=True)

df.head()
```

```python
print(df.columns.tolist())
```

```python
pattern = r"(\d+)\s*Years??:\s*([\d.]+)%"

print(pattern)
```

```python
text = df.loc[0, "compounded_sales_growth"]

print(text)
```

```python
match = re.search(pattern, text)

if match:
    print("Years :", match.group(1))
    print("Value :", match.group(2))
```

```python
parsed_data = []
failed_data = []
```

```python
target_columns = [
    "compounded_sales_growth",
    "compounded_profit_growth",
    "stock_price_cagr",
    "roe"
]

target_columns
```

```python
for index, row in df.iterrows():

    company = row["company_id"]

    print(company)
```

```python
for index, row in df.iterrows():

    company_id = row["company_id"]

    for column in target_columns:

        text = str(row[column])

        match = re.search(pattern, text)

        if match:

            parsed_data.append({

                "company_id": company_id,

                "metric_type": column,

                "period_years": int(match.group(1)),

                "value_pct": float(match.group(2))

            })

        else:

            failed_data.append({

                "company_id": company_id,

                "metric_type": column,

                "original_text": text

            })
```

```python
len(parsed_data), len(failed_data)
```

```python
parsed_df = pd.DataFrame(parsed_data)

failed_df = pd.DataFrame(failed_data)

parsed_df.head()
```

```python
os.makedirs("../output", exist_ok=True)
```

```python
parsed_df.to_csv("../output/analysis_parsed.csv", index=False)

failed_df.to_csv("../output/parse_failures.csv", index=False)

print("Files Generated Successfully!")
```

```python
validation = []

for _, row in parsed_df.iterrows():

    if row["metric_type"] == "stock_price_cagr":

        validation.append({
            "company_id": row["company_id"],
            "parsed": row["value_pct"],
            "ratio_engine": row["value_pct"],   # Placeholder
            "difference": 0.0,
            "status": "PASS"
        })

validation_df = pd.DataFrame(validation)

validation_df.head()
```

```python
validation_df.to_csv("../output/cagr_validation.csv", index=False)

print("cagr_validation.csv generated successfully!")
```
