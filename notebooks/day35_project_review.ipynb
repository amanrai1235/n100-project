```python
import pandas as pd
import os
from datetime import datetime
```

```python
files = [
    "../output/analysis_parsed.csv",
    "../output/parse_failures.csv",
    "../output/cagr_validation.csv",
    "../output/financial_metrics_standardized.csv",
    "../output/category_summary.csv",
    "../output/company_summary.csv",
    "../output/dashboard_kpis.csv",
    "../output/category_insights.csv",
    "../output/company_insights.csv",
    "../output/dashboard_company_dataset.csv",
    "../output/dashboard_category_dataset.csv",
    "../output/dashboard_summary.csv",
    "../reports/dashboard_testing_report.csv"
]

status = []

for file in files:
    status.append({
        "File": file,
        "Exists": os.path.exists(file)
    })

review_df = pd.DataFrame(status)

review_df
```

```python
completed = review_df["Exists"].sum()
total = len(review_df)

completion = round((completed / total) * 100, 2)

print("Completion:", completion, "%")
```

```python
summary = pd.DataFrame({
    "Project": ["N100 Financial Intelligence Platform"],
    "Sprint": ["Sprint 5"],
    "Completion (%)": [completion],
    "Status": [
        "Completed" if completion == 100 else "In Progress"
    ],
    "Generated_On": [
        datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    ]
})

summary
```

```python
os.makedirs("../reports", exist_ok=True)

review_df.to_csv(
    "../reports/final_review_report.csv",
    index=False
)

summary.to_csv(
    "../reports/sprint5_completion_report.csv",
    index=False
)

print("Final review reports generated successfully!")
```
