```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
portfolio = pd.read_csv("../reports/day22_portfolio.csv")

print(portfolio.shape)
print(portfolio.columns.tolist())

portfolio.head()
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    portfolio["company_id"],
    portfolio["allocation_percent"]
)

plt.title("Portfolio Allocation")
plt.xlabel("Company")
plt.ylabel("Allocation %")

plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("../reports/day24_allocation_chart.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    portfolio["company_id"],
    portfolio["composite_score"]
)

plt.title("Composite Score")

plt.xlabel("Company")

plt.ylabel("Score")

plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("../reports/day24_score_chart.png")

plt.show()
```

```python
risk = (
    portfolio["risk_category"]
    .value_counts()
)

plt.figure(figsize=(7,7))

plt.pie(
    risk.values,
    labels=risk.index,
    autopct="%1.1f%%"
)

plt.title("Risk Distribution")

plt.savefig("../reports/day24_risk_distribution.png")

plt.show()
```

```python
summary = pd.DataFrame({

    "Metric":[
        "Companies",
        "Average Score",
        "Average Allocation",
        "Highest Score",
        "Lowest Score"
    ],

    "Value":[
        len(portfolio),
        portfolio["composite_score"].mean(),
        portfolio["allocation_percent"].mean(),
        portfolio["composite_score"].max(),
        portfolio["composite_score"].min()
    ]

})

summary
```

```python
summary.to_csv(
    "../reports/day24_dashboard_summary.csv",
    index=False
)

print("Dashboard Summary Saved")
```
