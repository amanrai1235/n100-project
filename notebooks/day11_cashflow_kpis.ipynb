```python
cashflow = pd.read_sql("SELECT * FROM cashflow", conn)
```

```python
cashflow.columns = cashflow.iloc[0]
```

```python
cashflow = cashflow.iloc[1:].reset_index(drop=True)
```

```python
cashflow.head()
```

```python
pl = pd.read_sql("SELECT * FROM profitandloss", conn)

pl.columns = pl.iloc[0]

pl = pl.iloc[1:].reset_index(drop=True)

pl.head()
```

```python
df = cashflow.merge(
    pl,
    on=["company_id", "year"],
    how="inner"
)

df.head()
```

```python
numeric_cols = [
    "operating_activity",
    "investing_activity",
    "financing_activity",
    "operating_profit",
    "sales",
    "net_profit"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df.dtypes
```

```python
df["free_cash_flow"] = (
    df["operating_activity"] +
    df["investing_activity"]
)

df[
    [
        "company_id",
        "year",
        "operating_activity",
        "investing_activity",
        "free_cash_flow"
    ]
].head()
```

```python
import numpy as np

df["fcf_conversion_rate"] = np.where(
    df["operating_profit"] == 0,
    np.nan,
    (df["free_cash_flow"] / df["operating_profit"]) * 100
)

df[
    [
        "company_id",
        "year",
        "free_cash_flow",
        "operating_profit",
        "fcf_conversion_rate"
    ]
].head()
```

```python
df["capex_intensity"] = np.where(
    df["sales"] == 0,
    np.nan,
    (abs(df["investing_activity"]) / df["sales"]) * 100
)

df[
    [
        "company_id",
        "year",
        "sales",
        "investing_activity",
        "capex_intensity"
    ]
].head()
```

```python
df["cfo_pat_ratio"] = np.where(
    df["net_profit"] == 0,
    np.nan,
    df["operating_activity"] / df["net_profit"]
)

df[
    [
        "company_id",
        "year",
        "operating_activity",
        "net_profit",
        "cfo_pat_ratio"
    ]
].head()
```

```python
df["cfo_pat_avg_5yr"] = (
    df.groupby("company_id")["cfo_pat_ratio"]
      .transform(lambda x: x.rolling(5, min_periods=1).mean())
)

df[
    [
        "company_id",
        "year",
        "cfo_pat_ratio",
        "cfo_pat_avg_5yr"
    ]
].head(10)
```

```python
def cfo_quality(score):
    if pd.isna(score):
        return None
    elif score > 1:
        return "High Quality"
    elif score >= 0.5:
        return "Moderate"
    else:
        return "Accrual Risk"

df["cfo_quality"] = df["cfo_pat_avg_5yr"].apply(cfo_quality)

df[
    [
        "company_id",
        "year",
        "cfo_pat_avg_5yr",
        "cfo_quality"
    ]
].head(10)
```

```python
df["cfo_sign"] = np.where(df["operating_activity"] >= 0, "+", "-")
df["cfi_sign"] = np.where(df["investing_activity"] >= 0, "+", "-")
df["cff_sign"] = np.where(df["financing_activity"] >= 0, "+", "-")

df["pattern"] = (
    df["cfo_sign"] +
    df["cfi_sign"] +
    df["cff_sign"]
)

df[
    [
        "company_id",
        "year",
        "operating_activity",
        "investing_activity",
        "financing_activity",
        "pattern"
    ]
].head()
```

```python
pattern_labels = {
    "+--": "Reinvestor",
    "+-+": "Shareholder Returns",
    "++-": "Liquidating Assets",
    "-++": "Distress Signal",
    "--+": "Growth Funded by Debt",
    "+++": "Cash Accumulator",
    "---": "Pre-Revenue",
    "+-+": "Mixed"
}

df["capital_allocation"] = (
    df["pattern"]
      .map(pattern_labels)
      .fillna("Mixed")
)

df[
    [
        "company_id",
        "year",
        "pattern",
        "capital_allocation"
    ]
].head(20)
```

```python
df.to_csv(
    "../reports/capital_allocation_day11.csv",
    index=False
)

print("Saved Successfully")
```
