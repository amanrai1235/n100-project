```python
import pandas as pd
import numpy as np
import os
```

```python
parsed = pd.read_csv("../output/analysis_parsed.csv")

parsed.head()
```

```python
print(parsed.info())
print(parsed.isnull().sum())
```

```python
parsed = parsed.drop_duplicates()

print("Rows:", len(parsed))
```

```python
parsed["metric_type"] = (
    parsed["metric_type"]
    .str.strip()
    .str.lower()
)
```

```python
parsed["period_years"] = pd.to_numeric(
    parsed["period_years"],
    errors="coerce"
)

parsed["value_pct"] = pd.to_numeric(
    parsed["value_pct"],
    errors="coerce"
)
```

```python
parsed["period_years"] = parsed["period_years"].fillna(0)

parsed["value_pct"] = parsed["value_pct"].fillna(0)
```

```python
mapping = {
    "roe": "Profitability",
    "stock_price_cagr": "Returns",
    "compounded_sales_growth": "Growth",
    "compounded_profit_growth": "Growth"
}

parsed["metric_category"] = (
    parsed["metric_type"]
    .map(mapping)
)
```

```python
os.makedirs("../output", exist_ok=True)

parsed.to_csv(
    "../output/financial_metrics_standardized.csv",
    index=False
)

print("Financial metrics standardized successfully!")
```
