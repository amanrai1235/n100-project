```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect(
    "../database/nifty100.db"
)

print("Database Connected")
```

```python
companies = pd.read_csv(
    "../data/processed/companies_clean.csv"
)

analysis = pd.read_csv(
    "../data/processed/analysis_clean.csv"
)

balancesheet = pd.read_csv(
    "../data/processed/balancesheet_clean.csv"
)

cashflow = pd.read_csv(
    "../data/processed/cashflow_clean.csv"
)

profitandloss = pd.read_csv(
    "../data/processed/profitandloss_clean.csv"
)

documents = pd.read_csv(
    "../data/processed/documents_clean.csv"
)

prosandcons = pd.read_csv(
    "../data/processed/prosandcons_clean.csv"
)

print("Files Loaded")
```

```python
companies.to_sql(
    "companies",
    conn,
    if_exists="replace",
    index=False
)

analysis.to_sql(
    "analysis",
    conn,
    if_exists="replace",
    index=False
)

balancesheet.to_sql(
    "balancesheet",
    conn,
    if_exists="replace",
    index=False
)

cashflow.to_sql(
    "cashflow",
    conn,
    if_exists="replace",
    index=False
)

profitandloss.to_sql(
    "profitandloss",
    conn,
    if_exists="replace",
    index=False
)

documents.to_sql(
    "documents",
    conn,
    if_exists="replace",
    index=False
)

prosandcons.to_sql(
    "prosandcons",
    conn,
    if_exists="replace",
    index=False
)

print("Tables Created")
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table'
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS total_companies
FROM companies
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT *
FROM companies
LIMIT 10
"""

pd.read_sql(query, conn)
```
