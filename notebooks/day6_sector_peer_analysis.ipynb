```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sectors = pd.read_excel("../data/raw/supporting/sectors.xlsx")
peer_groups = pd.read_excel("../data/raw/supporting/peer_groups.xlsx")

print(sectors.shape)
print(peer_groups.shape)
```

```python
print(sectors.columns.tolist())

print(peer_groups.columns.tolist())
```

```python
sectors.head()
```

```python
peer_groups.head()
```

```python
print(sectors.isnull().sum())

print(peer_groups.isnull().sum())
```

```python
sector_count = sectors["broad_sector"].value_counts()

sector_count
```

```python
plt.figure(figsize=(10,5))

sector_count.plot(kind="bar")

plt.title("Companies by Sector")

plt.show()
```

```python
plt.figure(figsize=(8,8))

sector_count.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")
plt.title("Sector Distribution")

plt.show()
```

```python
sector_count.to_csv(
    "../reports/sector_distribution.csv"
)

print("Sector Report Saved")
```
