```python
import pandas as pd
import os
```

```python
company = pd.read_csv("../output/dashboard_company_dataset.csv")
category = pd.read_csv("../output/dashboard_category_dataset.csv")
summary = pd.read_csv("../output/dashboard_summary.csv")

print(company.head())
print(category.head())
print(summary.head())
```

```python
print("Company :", company.shape)
print("Category:", category.shape)
print("Summary :", summary.shape)
```

```python
print(company.isnull().sum())

print(category.isnull().sum())

print(summary.isnull().sum())
```

```python
print("Company duplicates :", company.duplicated().sum())

print("Category duplicates :", category.duplicated().sum())

print("Summary duplicates :", summary.duplicated().sum())
```

```python
testing_report = pd.DataFrame({

    "Dataset":[
        "Company",
        "Category",
        "Summary"
    ],

    "Rows":[
        len(company),
        len(category),
        len(summary)
    ],

    "Columns":[
        company.shape[1],
        category.shape[1],
        summary.shape[1]
    ],

    "Missing_Values":[
        company.isnull().sum().sum(),
        category.isnull().sum().sum(),
        summary.isnull().sum().sum()
    ]

})

testing_report
```

```python
os.makedirs("../reports", exist_ok=True)

testing_report.to_csv(
    "../reports/dashboard_testing_report.csv",
    index=False
)

print("Dashboard testing report generated.")
```
