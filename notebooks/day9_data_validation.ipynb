```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../database/nifty100.db")
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

pd.read_sql(query, conn)
```

```python
tables = [
    "companies",
    "balancesheet",
    "cashflow",
    "profitandloss",
    "analysis",
    "documents",
    "stock_prices",
    "financial_ratios",
    "peer_groups",
    "sectors"
]

for t in tables:
    q = f"SELECT COUNT(*) FROM {t}"
    print(t, pd.read_sql(q, conn))
```

```python
cursor = conn.cursor()

cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")

print(cursor.fetchall())
```

```python
tables = [
    "companies",
    "analysis",
    "balancesheet",
    "cashflow",
    "profitandloss",
    "documents",
    "prosandcons"
]

for t in tables:
    q = f"SELECT COUNT(*) FROM {t}"
    print("\nTable:", t)
    print(pd.read_sql(q, conn))
```

```python
for t in tables:
    print(f"\n===== {t} =====")

    df = pd.read_sql(f"SELECT * FROM {t}", conn)

    print(df.isnull().sum())
```

```python
for t in tables:
    print(f"\n===== {t} =====")

    df = pd.read_sql(f"SELECT * FROM {t}", conn)

    print("Duplicate Rows:", df.duplicated().sum())
```

```python
for t in tables:
    print(f"\n===== {t} =====")

    df = pd.read_sql(f"SELECT * FROM {t}", conn)

    print(df.dtypes)
```
