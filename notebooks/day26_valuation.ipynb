```python
analysis = pd.read_sql("SELECT * FROM analysis", conn)
analysis.columns = analysis.iloc[0]
analysis = analysis.iloc[1:]
analysis.head()
```

```python
profit = pd.read_sql("SELECT * FROM profitandloss", conn)
profit.columns = profit.iloc[0]
profit = profit.iloc[1:]
profit.head()
```

```python
balance = pd.read_sql("SELECT * FROM balancesheet", conn)
balance.columns = balance.iloc[0]
balance = balance.iloc[1:]
balance.head()
```

```python
valuation = profit.merge(
    analysis[["company_id", "roe"]],
    on="company_id",
    how="left"
)

valuation.head()
```

```python
valuation["roe"] = (
    valuation["roe"]
    .astype(str)
    .str.extract(r'(-?\d+\.?\d*)')[0]
    .astype(float)
)

valuation["opm_percentage"] = valuation["opm_percentage"].astype(float)
valuation["net_profit"] = valuation["net_profit"].astype(float)
```

```python
valuation["valuation_score"] = (
    valuation["roe"] * 0.4 +
    valuation["opm_percentage"] * 0.3 +
    valuation["net_profit"] * 0.3
)
```

```python
def rating(score):
    if score >= 100:
        return "Strong"
    elif score >= 50:
        return "Moderate"
    else:
        return "Weak"

valuation["valuation_rating"] = valuation["valuation_score"].apply(rating)
```

```python
final = valuation[
    [
        "company_id",
        "roe",
        "opm_percentage",
        "net_profit",
        "valuation_score",
        "valuation_rating"
    ]
]

final.head()
```

```python
final.to_csv("../reports/valuation_flags.csv", index=False)
```
