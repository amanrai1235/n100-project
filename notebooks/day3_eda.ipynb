```python
companies["roe_percentage"] = pd.to_numeric(
    companies["roe_percentage"],
    errors="coerce"
)

companies["roce_percentage"] = pd.to_numeric(
    companies["roce_percentage"],
    errors="coerce"
)
```

```python
top_roe = companies.sort_values(
    by="roe_percentage",
    ascending=False
)[["company_name","roe_percentage"]]

top_roe.head(10)
```

```python
top_roce = companies.sort_values(
    by="roce_percentage",
    ascending=False
)[["company_name","roce_percentage"]]

top_roce.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
top_roe.head(10).plot(
    x="company_name",
    y="roe_percentage",
    kind="bar",
    figsize=(10,5)
)

plt.title("Top 10 Companies by ROE")
plt.xticks(rotation=45)
plt.show()
```

```python
top_roce.head(10).plot(
    x="company_name",
    y="roce_percentage",
    kind="bar",
    figsize=(10,5)
)

plt.title("Top 10 Companies by ROCE")
plt.xticks(rotation=45)
plt.show()
```

```python
top_roe.head(10).to_csv(
    "../reports/top_roe_companies.csv",
    index=False
)

top_roce.head(10).to_csv(
    "../reports/top_roce_companies.csv",
    index=False
)

print("Reports Saved Successfully")
```
