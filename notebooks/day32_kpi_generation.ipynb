```python
import pandas as pd
import os
```

```python
category = pd.read_csv("../output/category_summary.csv")
company = pd.read_csv("../output/company_summary.csv")

category.head()
company.head()
```

```python
kpis = {
    "Total Companies": company["company_id"].nunique(),
    "Total Metrics": company["metrics_available"].sum(),
    "Average Financial Score": round(company["average_score"].mean(), 2),
    "Highest Financial Score": round(company["average_score"].max(), 2),
    "Lowest Financial Score": round(company["average_score"].min(), 2)
}

kpi_df = pd.DataFrame(
    kpis.items(),
    columns=["KPI", "Value"]
)

kpi_df
```

```python
category["Insight"] = category["avg_value"].apply(
    lambda x: "Strong" if x >= 20
    else "Moderate" if x >= 10
    else "Weak"
)

category
```

```python
company["Performance"] = company["average_score"].apply(
    lambda x: "Excellent" if x >= 20
    else "Good" if x >= 10
    else "Needs Improvement"
)

company.head()
```

```python
os.makedirs("../output", exist_ok=True)

kpi_df.to_csv("../output/dashboard_kpis.csv", index=False)

category.to_csv("../output/category_insights.csv", index=False)

company.to_csv("../output/company_insights.csv", index=False)

print("Day 32 outputs generated successfully!")
```
