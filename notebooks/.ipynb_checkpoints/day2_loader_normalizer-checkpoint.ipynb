```python
companies = pd.read_excel("../data/raw/companies.xlsx")
analysis = pd.read_excel("../data/raw/analysis.xlsx")
balancesheet = pd.read_excel("../data/raw/balancesheet.xlsx")
cashflow = pd.read_excel("../data/raw/cashflow.xlsx")
profitandloss = pd.read_excel("../data/raw/profitandloss.xlsx")
documents = pd.read_excel("../data/raw/documents.xlsx")
prosandcons = pd.read_excel("../data/raw/prosandcons.xlsx")

print("All files loaded successfully")
```

```python
datasets = {
    "companies": companies,
    "analysis": analysis,
    "balancesheet": balancesheet,
    "cashflow": cashflow,
    "profitandloss": profitandloss,
    "documents": documents,
    "prosandcons": prosandcons
}

for name, df in datasets.items():
    print(f"{name}: {df.shape}")
```

```python
for name, df in datasets.items():
    print("\n")
    print(name)
    print(df.columns.tolist())
```

```python
def normalize_column_names(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
    )
    return df
```

```python
companies = normalize_column_names(companies)
analysis = normalize_column_names(analysis)
balancesheet = normalize_column_names(balancesheet)
cashflow = normalize_column_names(cashflow)
profitandloss = normalize_column_names(profitandloss)
documents = normalize_column_names(documents)
prosandcons = normalize_column_names(prosandcons)
```

```python
companies.columns
```

```python
for name, df in datasets.items():
    print("\n")
    print(name)
    print(df.isnull().sum())
```

```python
for name, df in datasets.items():
    print(f"{name}: {df.duplicated().sum()} duplicate rows")
```

```python
import os

os.makedirs("../data/processed", exist_ok=True)

print("Processed folder ready")
```

```python
companies.to_csv("../data/processed/companies_clean.csv", index=False)
analysis.to_csv("../data/processed/analysis_clean.csv", index=False)
balancesheet.to_csv("../data/processed/balancesheet_clean.csv", index=False)
cashflow.to_csv("../data/processed/cashflow_clean.csv", index=False)
profitandloss.to_csv("../data/processed/profitandloss_clean.csv", index=False)
documents.to_csv("../data/processed/documents_clean.csv", index=False)
prosandcons.to_csv("../data/processed/prosandcons_clean.csv", index=False)

print("All cleaned files saved")
```
