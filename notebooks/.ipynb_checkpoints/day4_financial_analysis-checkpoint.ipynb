```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
financial_ratios = pd.read_excel("../data/raw/supporting/financial_ratios.xlsx")
market_cap = pd.read_excel("../data/raw/supporting/market_cap.xlsx")
peer_groups = pd.read_excel("../data/raw/supporting/peer_groups.xlsx")
sectors = pd.read_excel("../data/raw/supporting/sectors.xlsx")
stock_prices = pd.read_excel("../data/raw/supporting/stock_prices.xlsx")

print("All supporting datasets loaded")
```

```python
datasets = {
    "financial_ratios": financial_ratios,
    "market_cap": market_cap,
    "peer_groups": peer_groups,
    "sectors": sectors,
    "stock_prices": stock_prices
}

for name, df in datasets.items():
    print(name, df.shape)
```

```python
for name, df in datasets.items():
    print("\n")
    print(name)
    print(df.columns.tolist())
```

```python
top_roe = (
    financial_ratios
    .sort_values("return_on_equity_pct", ascending=False)
)

top_roe[["company_id","return_on_equity_pct"]].head(10)
```

```python
top_roe_chart = top_roe[["company_id","return_on_equity_pct"]].head(10)

plt.figure(figsize=(10,5))
plt.bar(
    top_roe_chart["company_id"],
    top_roe_chart["return_on_equity_pct"]
)

plt.title("Top 10 Companies by ROE")
plt.xticks(rotation=45)
plt.show()
```

```python
top_market_cap = (
    market_cap
    .sort_values("market_cap_crore", ascending=False)
)

top_market_cap[
    ["company_id","market_cap_crore"]
].head(10)
```

```python
top_market_chart = top_market_cap[
    ["company_id","market_cap_crore"]
].head(10)

plt.figure(figsize=(10,5))
plt.bar(
    top_market_chart["company_id"],
    top_market_chart["market_cap_crore"]
)

plt.title("Top 10 Companies by Market Cap")
plt.xticks(rotation=45)
plt.show()
```

```python
sector_count = sectors["broad_sector"].value_counts()

sector_count
```

```python
plt.figure(figsize=(8,8))

sector_count.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")
plt.title("Sector Distribution")
plt.show()
```

```python
top_roe[[
    "company_id",
    "return_on_equity_pct"
]].head(10).to_csv(
    "../reports/top_roe_companies.csv",
    index=False
)

top_market_cap[[
    "company_id",
    "market_cap_crore"
]].head(10).to_csv(
    "../reports/top_market_cap_companies.csv",
    index=False
)

print("Reports Saved Successfully")
```
